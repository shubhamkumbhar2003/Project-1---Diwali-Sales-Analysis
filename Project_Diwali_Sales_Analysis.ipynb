```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(r'Diwali Sales Data.csv' , encoding = 'unicode_escape')
```

```python
df.shape#it will tell (rows columns) present in the data
```

```python
#for the Data Cleaning 
df.info()
```

```python
df.drop(['Status' , 'unnamed1']  , axis = 1 , inplace = True)
```

```python
df.info()
```

```python
#For checking the null values present in the csv file
pd.isnull(df).sum()
```

```python
#here in amount column there are 12 such rows which has the null values
#to delete the null values
df.dropna(inplace=True)
```

```python
df.shape #shows the row and columns which are present now in the csv file
```

```python
df.head(10)
```

```python
df.info()
```

```python
#Changing the datatype
df['Amount'] = df['Amount'].astype('int')
```

```python
#To change the name of the column
df.rename(columns={'Cust_name':'Customer_name'} , inplace = True)
```

```python
df.columns
```

```python
axis = sns.countplot(x = 'Gender' , data = df)
for i in axis.containers:
    axis.bar_label(i)
```

Here in above countplot we have analyzed the Count of the Female and the male in file

```python
sales_gender = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = True)
```

```python
sales_gender
```

```python
bar = sns.barplot(x = 'Gender' , y='Amount' , data = sales_gender)
for i in bar.containers:
    bar.bar_label(i)
```

Here in above barplot we have analyzed the amount spent by each male and the female gender

Age

```python
df.columns
```

```python
bar = sns.countplot(x='Age Group' , data =df , hue = 'Gender' , )
for i in bar.containers:
    bar.bar_label(i)
```

The above countplot shows us that the largest age group people are in between 26-35 among them females are the most in the count

```python
sales_age = df.groupby(['Age Group','Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False)
bar = sns.barplot(x = 'Age Group' , y = 'Amount' , data = sales_age , hue = 'Gender')
    
```

In the above barplot we can see clearly that the age-group 26-35 has spent more amount and in that age-group female has spent the most

States

```python
df.columns
```

```python
sales_state = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders' , ascending = False).head(10)
sns.set(rc = {'figure.figsize':(15,5)})
sns.barplot(x = 'State' , y = 'Orders' ,data = sales_state)
```

In the above barplot, the maximum orders are from the uttar Pradesh state

Marital Status

```python
ax = sns.countplot(x = 'Marital_Status' , data = df)
sns.set(rc = {'figure.figsize':(7,9)})
for i in ax.containers:
    ax.bar_label(i)
```

```python
total_data = df.groupby(['Marital_Status' , 'Gender'] , as_index = False)['Amount'].sum().sort_values(by='Amount' , ascending = False)
sns.barplot(data = total_data , x = 'Marital_Status' , y = 'Amount' , hue = 'Gender')
```

Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(x = 'Occupation' , data = df)
for i in ax.containers:
    ax.bar_label(i)
```

In all the working sectors the most widely working sector is the IT Sector that has nearly 1583 counts.

```python
sales_state = df.groupby(['Occupation' , 'Gender'] , as_index =False)['Amount'].sum().sort_values(by = 'Amount' , ascending = True)
sns.barplot(x = 'Occupation' , y = 'Amount' , data = sales_state , hue = 'Gender')
```

As seen, the working people are the most in the IT Sector and yes we can also see that Females are the most in the IT Sector as compare to Male

Product Category

```python
sns.set(rc = {'figure.figsize':(25,5)})
ax = sns.countplot(x = 'Product_Category' , data = df)
for i in ax.containers:
    ax.bar_label(i)
```

In the above countplot, the most purchased Product Category is the clothing & apparel

```python
sales_Product = df.groupby(['Product_Category'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = True)
sns.set(rc = {'figure.figsize':(27,5)})
sns.barplot(x = 'Product_Category' , y = 'Amount' , data = sales_Product)
```

In the above barplot the most amount spend on the product category is the food

```python
sales_Orders = df.groupby(['Product_ID'] , as_index = False)['Orders'].sum().sort_values(by ='Orders' , ascending = False).head(10)
sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(x = 'Product_ID' , y = 'Orders' , data = sales_Orders)
```
